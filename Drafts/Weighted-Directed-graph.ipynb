```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a directed graph
G = nx.DiGraph()

# Add nodes. For example, let's add 4 nodes.
G.add_node("A")
G.add_node("B")
G.add_node("C")
G.add_node("D")

# Add edges with weights
G.add_edge("A", "B", weight=1.5)
G.add_edge("A", "C", weight=2.5)
G.add_edge("B", "C", weight=2.0)
G.add_edge("C", "D", weight=3.0)

# Draw the graph
pos = nx.spring_layout(G)  # positions for all nodes

# nodes
nx.draw_networkx_nodes(G, pos, node_size=700)

# edges
nx.draw_networkx_edges(G, pos, edgelist=G.edges(data=True), arrowstyle='->', arrowsize=20)

# labels
nx.draw_networkx_labels(G, pos, font_size=20, font_family="sans-serif")

# edge weight labels
edge_labels = nx.get_edge_attributes(G, "weight")
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)

plt.axis("off")
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a directed graph
G = nx.DiGraph()

# Add nodes
G.add_node("A")
G.add_node("B")
G.add_node("C")
G.add_node("D")

# Add weighted edges
G.add_edge("A", "B", weight=1.5)
G.add_edge("A", "C", weight=2.5)
G.add_edge("B", "C", weight=2.0)
G.add_edge("C", "D", weight=3.0)

# Compute centrality measures
degree_centrality = nx.degree_centrality(G)
closeness_centrality = nx.closeness_centrality(G, distance='weight')
betweenness_centrality = nx.betweenness_centrality(G, weight='weight')
eigenvector_centrality = nx.eigenvector_centrality(G, weight='weight', max_iter=1000)
pagerank = nx.pagerank(G, weight='weight')

# Print centrality measures
print("Degree Centrality:", degree_centrality)
print("Closeness Centrality:", closeness_centrality)
print("Betweenness Centrality:", betweenness_centrality)
print("Eigenvector Centrality:", eigenvector_centrality)
print("PageRank:", pagerank)

# Draw the graph
pos = nx.spring_layout(G)  # positions for all nodes

# nodes
nx.draw_networkx_nodes(G, pos, node_size=700)

# edges
weights = [G[u][v]['weight'] for u,v in G.edges()]
nx.draw_networkx_edges(G, pos, edgelist=G.edges(), width=weights, arrowstyle='->', arrowsize=20)

# labels
nx.draw_networkx_labels(G, pos, font_size=20, font_family="sans-serif")

# edge weight labels
edge_labels = nx.get_edge_attributes(G, "weight")
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)

plt.axis("off")
plt.show()
```

```python
import networkx as nx

# Create a directed graph
G = nx.DiGraph()

# Add nodes
G.add_node("A")
G.add_node("B")
G.add_node("C")
G.add_node("D")

# Add weighted edges
G.add_edge("A", "B", weight=1.5)
G.add_edge("A", "C", weight=2.5)
G.add_edge("B", "C", weight=2.0)
G.add_edge("C", "D", weight=3.0)

# Function to calculate weighted in-degree centrality
def weighted_in_degree_centrality(G):
    in_degrees = {node: 0 for node in G}
    for u, v, data in G.in_edges(data=True):
        in_degrees[v] += data.get('weight', 1)  # default to 1 if no weight
    max_in_degree = max(in_degrees.values()) if in_degrees else 1
    return {node: deg / max_in_degree for node, deg in in_degrees.items()}

# Function to calculate weighted out-degree centrality
def weighted_out_degree_centrality(G):
    out_degrees = {node: 0 for node in G}
    for u, v, data in G.out_edges(data=True):
        out_degrees[u] += data.get('weight', 1)  # default to 1 if no weight
    max_out_degree = max(out_degrees.values()) if out_degrees else 1
    return {node: deg / max_out_degree for node, deg in out_degrees.items()}

# Calculate weighted in-degree and out-degree centralities
weighted_in_centrality = weighted_in_degree_centrality(G)
weighted_out_centrality = weighted_out_degree_centrality(G)

# Print the results
print("Weighted In-Degree Centrality:", weighted_in_centrality)
print("Weighted Out-Degree Centrality:", weighted_out_centrality)
```

```python
import networkx as nx

# Create a directed graph
G = nx.DiGraph()

# Add nodes
G.add_node("A")
G.add_node("B")
G.add_node("C")
G.add_node("D")

# Add weighted edges
G.add_edge("A", "B", weight=1.5)
G.add_edge("A", "C", weight=2.5)
G.add_edge("B", "C", weight=2.0)
G.add_edge("C", "D", weight=3.0)

# Function to calculate weighted in-degree centrality
def weighted_in_degree_centrality(G):
    in_degrees = {node: 0 for node in G}
    total_in_weight = 0
    for u, v, data in G.in_edges(data=True):
        weight = data.get('weight', 1)  # default to 1 if no weight
        in_degrees[v] += weight
        total_in_weight += weight
    return {node: deg / total_in_weight for node, deg in in_degrees.items()}

# Function to calculate weighted out-degree centrality
def weighted_out_degree_centrality(G):
    out_degrees = {node: 0 for node in G}
    total_out_weight = 0
    for u, v, data in G.out_edges(data=True):
        weight = data.get('weight', 1)  # default to 1 if no weight
        out_degrees[u] += weight
        total_out_weight += weight
    return {node: deg / total_out_weight for node, deg in out_degrees.items()}

# Calculate weighted in-degree and out-degree centralities
weighted_in_centrality = weighted_in_degree_centrality(G)
weighted_out_centrality = weighted_out_degree_centrality(G)

# Print the results
print("Weighted In-Degree Centrality:", weighted_in_centrality)
print("Weighted Out-Degree Centrality:", weighted_out_centrality)
```
